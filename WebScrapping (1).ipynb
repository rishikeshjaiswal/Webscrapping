```python
final=pd.DataFrame()
for j in range(1,10):
    url='https://www.ambitionbox.com/list-of-companies?campaign=desktop_nav&page={}'.format(j)
    header={"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
                  "AppleWebKit/537.36 (KHTML, like Gecko) "
                  "Chrome/120.0.0.0 Safari/537.36"}
    web=requests.get(url,headers=header).text
    soup=BeautifulSoup(web,'lxml')
    company=soup.find_all('div',class_='companyCardWrapper')
    name=[]
    rating=[]
    nreviews=[]
    sl=[]
    salary=[]
    job=[]
    benefit=[]
    
    for i in company:
        name.append(i.find('h2').text.strip())
        rating.append(i.find('div', class_='rating_text').text.strip())
        nreviews.append(i.find('span', class_='companyCardWrapper__companyRatingCount').text.strip())
        sl.append(i.find('span', class_='companyCardWrapper__interLinking').text.strip())
        salary.append(i.find_all('span',class_='companyCardWrapper__ActionCount')[1].text.strip())
        job.append(i.find_all('span',class_='companyCardWrapper__ActionCount')[3].text.strip())
        benefit.append(i.find_all('span',class_='companyCardWrapper__ActionCount')[4].text.strip())
        
    df={'Name':name,'Rating': rating, 'Total reviews': nreviews,'Sevice & Location': sl,'Salary': salary,
        'Jobs': job,'Benefits': benefit}
    df1=pd.DataFrame(df)
    final = pd.concat([final, df1], ignore_index=True)
```
